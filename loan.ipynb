```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import pickle
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
```

```python
df.dropna(inplace = True)
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.loan_id.unique()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df['self_employed'] = df['self_employed'].str.strip().replace({'Yes':1, 'No': 0})
```

```python
df
```

```python
df['education'] = df['education'].str.strip().replace({'Graduate':1, 'Not Graduate': 0})
```

```python
df
```

```python
df['loan_status'] = df['loan_status'].str.strip().replace({'Approved':1, 'Rejected': 0})
```

```python
df
```

```python
X = df.drop(columns = ['residential_assets_value','commercial_assets_value','luxury_assets_value', 'loan_status'])
y = df['loan_status']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.33, random_state = 42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
